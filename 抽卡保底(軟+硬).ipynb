```python
#  Ëªü‰øùÂ∫ïÊäΩÂç°ÊúüÊúõÂÄºÊ®°Êì¨

import matplotlib.pyplot as plt

# ÊäΩÂç°Ë®≠ÂÆöÂèÉÊï∏
n = 100  # Á∏ΩÊäΩÊï∏
p = 0.005  # ÂàùÂßãSSRÊ©üÁéá(0.5%)
soft_pity_start = 75  # Ëªü‰øùÂ∫ïÈñãÂßã
hard_pity = 100  # Á°¨‰øùÂ∫ï
pity_increment = 0.03  # ÊØèÊäΩÊ©üÁéáÂ¢ûÂä† 3%

# ÊØèÊäΩ‰∏≠SSRÁöÑÊ©üÁéá
draw_probs = []
for i in range(1, n + 1):
    if i < soft_pity_start:
        draw_probs.append(p)
    elif i < hard_pity:
        inc = (i - soft_pity_start + 1) * pity_increment
        draw_probs.append(min(p + inc, 1.0))
    else:
        draw_probs.append(1.0)

# ÊØèÊäΩ„ÄåÁ¨¨‰∏ÄÊ¨°‰∏≠SSR„ÄçÁöÑÊ©üÁéáÔºàÈùûÁ¥ØÁ©çÊ©üÁéáÔºâ

first_hit_probs = []
cum_prob = 1.0
for prob in draw_probs:
    first = prob * cum_prob
    first_hit_probs.append(first)
    cum_prob *= (1 - prob)


# Ë®àÁÆóÊúüÊúõÂÄº
expected_draws = sum([i * prob for i, prob in enumerate(first_hit_probs, start=1)])

print(f"üéØ Ê®°ÂûãÊúüÊúõÊäΩÊï∏ÔºöÁ¥Ñ {expected_draws:.2f} ÊäΩÂèØ‰∏≠ 1 SSR")

# Áï´ÂúñË¶ñË¶∫Âåñ
plt.figure(figsize=(14, 7))
plt.plot(range(1, n + 1), draw_probs, label="ÊØèÊäΩ‰∏≠SSRÊ©üÁéá")
plt.plot(range(1, n + 1), first_hit_probs, label="ÊØèÊäΩÈ¶ñÊ¨°‰∏≠SSRÊ©üÁéá")
plt.axvline(expected_draws, color='r', linestyle='--', label=f"ÊúüÊúõÊäΩÊï∏ ‚âà {expected_draws:.2f}")
plt.title("Ëªü‰øùÂ∫ïÊäΩÂç°Ê©üÁéáÂàÜÊûê")
plt.xlabel("Á¨¨ÂπæÊäΩ")
plt.ylabel("Ê©üÁéá")
plt.legend()
plt.grid(True)
plt.show()
```
